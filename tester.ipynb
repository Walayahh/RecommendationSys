```python
import numpy as np 
import csv
import pandas as pd
```

```python
perfumes_df = pd.read_csv("perfumes.csv", delimiter=";", encoding="latin-1")


# Define a function to replace dashes with spaces and convert to title case
def transform(text):
    if isinstance(text, str):
        return text.replace("-", " ").title()
    return text

# Apply the transformation to both the 'Perfume' and 'Brand' columns
perfumes_df["Perfume"] = perfumes_df["Perfume"].apply(transform)
perfumes_df["Brand"] = perfumes_df["Brand"].apply(transform)
perfumes_df["Perfumes"] = perfumes_df["Brand"].apply(transform) + " " +  perfumes_df["Perfume"].apply(transform)
users_df = pd.read_csv("users.csv")
```

### Non-Personalized Recommendations

```python

# Check that the column exists and see its content
print("Columns in DataFrame:", users_df.columns)
print("First few rows:")
print(users_df.head())

# Assuming the column with comma-separated perfumes is named "perfumes"
# Create a new column 'perfume_list' by splitting each string on commas
users_df['perfume_list'] = users_df['Perfumes'].apply(
    lambda x: [p.strip() for p in x.split(',')] if isinstance(x, str) else []
)

# Verify the new column
print("After splitting, sample data:")
print(users_df[['Perfumes', 'perfume_list']].head())

# Explode the lists so each perfume is on its own row
exploded_df = users_df.explode('perfume_list')

# Verify the exploded DataFrame
print("After exploding:")
print(exploded_df[['perfume_list']].head(10))

# Count the number of times each perfume was selected
perfume_counts = exploded_df['perfume_list'].value_counts().reset_index()
perfume_counts.columns = ['perfume', 'count']

# Display the top entries
print("Perfume counts:")
perfume_counts.head()
perfume_counts.to_csv("perfume_counts.csv", index=False)
```
