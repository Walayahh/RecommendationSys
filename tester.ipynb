```python
# Alwaleed Item based recommendations
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import logging

# Configure a basic logger.
# - level=logging.DEBUG will give very detailed logs.
# - level=logging.INFO will be less verbose.
# The format can include timestamps, logging level, and the message.
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s [%(levelname)s] %(name)s - %(message)s'
)
logger = logging.getLogger(__name__)


###############################################################################
# 1) CONTENT-BASED SIMILARITY
###############################################################################
def prepare_content_based_similarity(df_perfumes):
    """
    Given a DataFrame of perfumes (df_perfumes), each row containing textual data
    such as Top, Middle, Base notes, main accords, and perfumers,
    compute a content-based similarity matrix using TF-IDF + cosine similarity.

    Returns a DataFrame content_sim with shape [n_perfumes, n_perfumes],
    whose index and columns both match the 'Perfume' column of df_perfumes.
    """
    # Combine relevant textual fields into one big string for each perfume.
    # This is typical of content-based text processing with TF-IDF.
    df_perfumes['combined_features'] = (
        df_perfumes['Top'].fillna('') + ' ' +
        df_perfumes['Middle'].fillna('') + ' ' +
        df_perfumes['Base'].fillna('') + ' ' +
        df_perfumes['mainaccord1'].fillna('') + ' ' +
        df_perfumes['mainaccord2'].fillna('') + ' ' +
        df_perfumes['mainaccord3'].fillna('') + ' ' +
        df_perfumes['mainaccord4'].fillna('') + ' ' +
        df_perfumes['mainaccord5'].fillna('') + ' ' +
        df_perfumes['Perfumer1'].fillna('') + ' ' +
        df_perfumes['Perfumer2'].fillna('')
    )

    # Use TF-IDF to vectorize the combined text fields.
    # stop_words='english' removes common words that typically aren’t helpful for similarity.
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(df_perfumes['combined_features'])

    # Compute cosine similarity between all perfume vectors.
    content_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Create a DataFrame with index & columns both set to the perfume names (must be unique).
    content_sim = pd.DataFrame(
        content_sim_matrix,
        index=df_perfumes['Perfume'],
        columns=df_perfumes['Perfume']
    )
    return content_sim


###############################################################################
# 2) ITEM-BASED COLLABORATIVE FILTERING SIMILARITY
###############################################################################
def prepare_item_based_similarity(df_users, df_perfumes):
    """
    Given:
      - df_users: user-level data, each row representing one user’s selections,
                  with a 'Perfumes' column (comma-separated) or similar approach.
      - df_perfumes: a DataFrame of all possible perfumes, with a 'Perfume' column.

    Creates a user x item matrix (0/1), transposes to item x user, and computes an
    item-based collaborative similarity matrix (collab_sim) using cosine similarity.

    Returns a DataFrame collab_sim with shape [n_perfumes, n_perfumes].
    """
    # Get all perfume names from the main df_perfumes.
    all_perfume_names = df_perfumes['Perfume'].unique().tolist()

    # Build a map: user_id -> set_of_perfumes (in lowercase) the user has chosen.
    user_perfume_map = {}
    for idx, row in df_users.iterrows():
        # If 'Perfumes' is a comma-separated list of user-chosen perfumes:
        raw_list = str(row['Perfumes']).split(',')
        chosen = set([p.strip().lower() for p in raw_list if p.strip()])
        user_perfume_map[idx] = chosen

    # Create a user-item matrix with shape [n_users, n_perfumes], fill with 0 initially.
    user_ids = list(user_perfume_map.keys())
    user_item_matrix = pd.DataFrame(
        data=0,
        index=user_ids,
        columns=all_perfume_names
    )

    # For faster name matching: map each perfume's lowercase name -> original name
    name_map = {p.lower(): p for p in all_perfume_names}

    # Fill 1 if the user selected that perfume.
    for user_id, chosen_perfumes in user_perfume_map.items():
        for cp in chosen_perfumes:
            if cp in name_map:
                real_name = name_map[cp]
                user_item_matrix.at[user_id, real_name] = 1

    # Transpose to get item x user.
    item_user_matrix = user_item_matrix.T  # shape: [n_perfumes, n_users]

    # Compute item-to-item (perfume-to-perfume) similarity using cosine similarity.
    collab_sim_matrix = cosine_similarity(item_user_matrix, item_user_matrix)

    # Convert to a DataFrame with index/columns = the actual perfume names.
    collab_sim = pd.DataFrame(
        collab_sim_matrix,
        index=all_perfume_names,
        columns=all_perfume_names
    )
    return collab_sim


###############################################################################
# 3) COMBINING CONTENT & COLLABORATIVE SIMILARITY
###############################################################################
def prepare_combined_similarity(content_sim, collab_sim, alpha=0.5):
    """
    Combine content-based (content_sim) and collaborative (collab_sim) similarity
    matrices with a weighting factor alpha.

    combined_sim = alpha * collab_sim + (1 - alpha) * content_sim

    Both content_sim and collab_sim are DataFrames with the same shape
    and same index/columns (perfume names).
    """
    # Identify the intersection of perfume names that exist in both matrices.
    common_perfumes = content_sim.index.intersection(collab_sim.index)

    # Reindex them so they have the same row and column ordering.
    content_sim = content_sim.loc[common_perfumes, common_perfumes]
    collab_sim = collab_sim.loc[common_perfumes, common_perfumes]

    # Weighted sum to combine the two similarities.
    combined = alpha * collab_sim + (1 - alpha) * content_sim
    return combined


###############################################################################
# 4) RECOMMENDATION FUNCTION
###############################################################################
def get_item_based_recommendations(selected_perfumes, combined_sim, top_n=5):
    """
    Given:
      - selected_perfumes: a list of perfume names the user already likes/has chosen
      - combined_sim: a DataFrame (perfume x perfume) with combined similarity
      - top_n: number of recommendations to return

    We aggregate similarity to each perfume from all of the user’s selected items,
    then sort by descending similarity. We exclude any perfumes the user already selected.
    Returns a list of recommended perfume names.
    """
    agg_scores = {}
    all_perfumes = combined_sim.index.tolist()

    # Build a quick map from lower-case to the original perfume name for robust matching.
    lower_to_real = {p.lower(): p for p in all_perfumes}

    # Convert user-selected perfumes to lower-case for matching.
    selected_lower = [p.lower() for p in selected_perfumes if isinstance(p, str)]

    # For each selected perfume, accumulate the similarity scores to all other perfumes.
    for p in selected_lower:
        if p in lower_to_real:
            real_p = lower_to_real[p]
            # This row is a Series of similarities from real_p to every other perfume.
            row_sim = combined_sim.loc[real_p]
            for perfume_name, sim_val in row_sim.items():
                # Skip if it's one of the user’s already selected items
                if perfume_name.lower() in selected_lower:
                    continue
                agg_scores[perfume_name] = agg_scores.get(perfume_name, 0) + sim_val

    # Sort aggregated scores in descending order by similarity sum.
    sorted_perfumes = sorted(agg_scores.items(), key=lambda x: x[1], reverse=True)

    # Return the top_n perfume names.
    top_recs = [perf for perf, score in sorted_perfumes[:top_n]]
    return top_recs


###############################################################################
# EXAMPLE / DEMO
###############################################################################
if __name__ == '__main__':
    # 1) Read your content-based data from perfumes.csv (semicolon-delimited).
    df_perfumes = perfumes_df
    # Make sure df_perfumes has columns like 'Perfume', 'Top', 'Middle', 'Base',
    # 'mainaccord1', 'Perfumer1', etc. (Adjust as needed.)

    # 2) Read user data from users.csv (comma-delimited).
    #    Suppose it has columns: 'Perfumes' (with comma-separated perfume picks), 'Gender', etc.
    df_users = pd.read_csv('Users.csv', delimiter=',')

    # 3) Compute content-based similarity using TF-IDF from your content-based notebook approach.
    content_sim = prepare_content_based_similarity(df_perfumes)

    # 4) Compute item-based collaborative similarity from the user-likes data.
    collab_sim = prepare_item_based_similarity(df_users, df_perfumes)

    # 5) Merge them with some weighting alpha (0.0 -> purely content-based,
    #    1.0 -> purely collaborative; something in between -> a hybrid).
    alpha = 0.65
    combined_sim = prepare_combined_similarity(content_sim, collab_sim, alpha)

    # 6) Suppose the user selected these perfumes:
    user_selected = [
        "Mercedes Benz Mercedes Benz L Eau, Zara Batman",
        "Charriol Charriol Homme Sport",
        "Givenchy Ange Ou Demon 10 Years",
        "Puma Green",
        "Replay Stone Supernova For Him",
        "Mauboussin Mauboussin A La Folie",
        "Estee Lauder Metropolis",
        "Mancera Aoud Exclusif",
        "Mdci Parfums Vepres Siciliennes"
    ]

    # 7) Get top 5 recommendations
    top_n = 5
    recommendations = get_item_based_recommendations(user_selected, combined_sim, top_n)
    print("User selected:", user_selected)
    print(f"Hybrid recommendations (alpha={alpha}) =>", recommendations)

```

```python
#############################################
# rowwise_content_recommender.py
#############################################
"""
Implements a row-by-row top-k approach for content-based recommendations
using TF-IDF, to avoid constructing a huge NxN similarity matrix.

Flow:
  1) Read perfumes.csv, combine textual features.
  2) Build TF-IDF: shape (N, M).
  3) For each user-selected perfume:
     a) Multiply that perfume's TF-IDF row vector by tfidf_matrix.T (N x 1).
     b) Get top_k neighbors from that single row (excl. itself).
     c) Combine (sum) scores if multiple items are selected.
  4) Return final top-n recommendations.
No MemoryError from NxN because we never do a full 'cosine_similarity(tfidf, tfidf)'.
"""

import logging
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from heapq import nlargest

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(name)s - %(message)s'
)
logger = logging.getLogger(__name__)

#############################################
# 1) READ & PREPARE THE DATA
#############################################
def load_and_prepare_perfumes(csv_path, delimiter=";", encoding="latin-1"):
    """
    Loads your 'perfumes.csv', combines textual columns (Top, Middle, Base, etc.)
    into a single 'combined_features' column, and returns the DataFrame.
    Adjust the column names as needed to match your actual CSV.
    """
    logger.info("Loading perfumes data from %s", csv_path)
    df = pd.read_csv(csv_path, delimiter=delimiter, encoding=encoding)

    # Ensure we have columns: 'Top', 'Middle', 'Base', 'mainaccord1'...'mainaccord5', etc.
    # If you have a different structure, adjust accordingly.
    df["combined_features"] = (
        df["Top"].fillna("") + " " +
        df["Middle"].fillna("") + " " +
        df["Base"].fillna("") + " " +
        df["mainaccord1"].fillna("") + " " +
        df["mainaccord2"].fillna("") + " " +
        df["mainaccord3"].fillna("") + " " +
        df["mainaccord4"].fillna("") + " " +
        df["mainaccord5"].fillna("")
        # If you want to include Perfumer or Brand, add them here
    )

    # Possibly rename 'Perfume' column if you want or unify naming
    logger.info("Perfumes loaded with shape: %s", df.shape)
    return df

#############################################
# 2) TF-IDF and Indices
#############################################
def build_tfidf_matrix(df, text_col="combined_features"):
    """
    Build a TF-IDF matrix from the DataFrame's 'text_col' column.
    Returns:
      - tfidf_matrix (scipy sparse NxM)
      - index_to_name (list of perfume names)
      - name_to_index (dict for lookup)
    """
    documents = df[text_col].fillna("").values
    logger.info("Building TF-IDF for %d documents...", len(documents))

    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform(documents)
    logger.info("TF-IDF matrix shape: %s", tfidf_matrix.shape)

    # Build index <-> perfume name mapping
    index_to_name = df["Perfume"].tolist()
    name_to_index = {name: i for i, name in enumerate(index_to_name)}

    return tfidf_matrix, index_to_name, name_to_index

#############################################
# 3) ROW-BY-ROW TOP-K FUNCTION
#############################################
def get_topk_for_perfume(
    perfume_name,
    tfidf_matrix,
    name_to_index,
    index_to_name,
    top_k=50
):
    """
    Given a single 'perfume_name', compute similarity row:
      row = tfidf_matrix[row_i] dot tfidf_matrix.T => shape (N,)
    Then pick top_k results (excluding itself).
    Returns a list of (perfume_name, score).
    """
    perfume_name_lower = perfume_name.lower()
    # 1) Find the index
    # We do a direct match. If there's a mismatch in naming, we'll skip
    matched_index = None
    for real_name, i in name_to_index.items():
        if real_name.lower() == perfume_name_lower:
            matched_index = i
            break
    if matched_index is None:
        logger.warning("Perfume '%s' not found in name_to_index. Returning empty.", perfume_name)
        return []

    # 2) Dot product row
    # row_i => shape (1, M)
    row_i = tfidf_matrix[matched_index]
    # sim_vec => shape (1, N) after row_i * tfidf_matrix^T => we get dense or sparse?
    # We'll do .toarray() so we can manipulate it easily, but we must watch memory usage.
    sim_vec = row_i.dot(tfidf_matrix.T).toarray().ravel()  # shape (N,)

    # 3) Exclude itself
    sim_vec[matched_index] = -999.0

    # 4) find top_k via partial sort
    # We'll gather the top_k indices, then create a list (perf_name, score).
    if top_k >= len(sim_vec):
        # if top_k is bigger than the dataset, just sort the entire array
        top_indices = np.argsort(sim_vec)[::-1][:top_k]
    else:
        # partial selection
        # or we can do 'np.argpartition(sim_vec, -top_k)' if we want partial
        # for simplicity, let's do that
        partition_idx = np.argpartition(sim_vec, -top_k)[-top_k:]
        top_indices = partition_idx[np.argsort(sim_vec[partition_idx])[::-1]]

    results = [(index_to_name[idx], sim_vec[idx]) for idx in top_indices]
    return results

#############################################
# 4) AGGREGATE MULTIPLE SELECTED PERFUMES
#############################################
def get_content_based_recommendations(
    selected_perfumes,
    tfidf_matrix,
    name_to_index,
    index_to_name,
    top_k_each=50,
    final_n=5
):
    """
    If the user selected multiple perfumes, we:
      - For each selected perfume, get top_k_each neighbors.
      - Sum up similarity scores across those sets.
      - Sort & pick the final_n best suggestions (excluding the user selected).
    Returns a list of recommended perfume names.
    """
    logger.info("Starting content-based rowwise aggregator for %s", selected_perfumes)
    agg_scores = {}
    selected_lower = {p.lower() for p in selected_perfumes}

    for perfume in selected_perfumes:
        logger.info("Fetching top-%d for '%s'...", top_k_each, perfume)
        topk_list = get_topk_for_perfume(
            perfume,
            tfidf_matrix,
            name_to_index,
            index_to_name,
            top_k=top_k_each
        )
        # topk_list => [(name, score), (name2, score2), ...]
        for (name, score) in topk_list:
            if name.lower() in selected_lower:
                # skip if user already selected it
                continue
            agg_scores[name] = agg_scores.get(name, 0.0) + score

    if not agg_scores:
        logger.warning("No scores aggregated. Possibly no matches or mismatch in naming.")
        return []

    # Sort by descending aggregated score
    sorted_items = sorted(agg_scores.items(), key=lambda x: x[1], reverse=True)
    final_recs = [name for (name, score) in sorted_items[:final_n]]
    logger.info("Final recommendations => %s", final_recs)
    return final_recs

#############################################
# 5) DEMO
#############################################
if __name__ == "__main__":
    # A) Load the data
    perfumes_df = load_and_prepare_perfumes("perfumes_updated.csv")
    # B) Build TF-IDF
    tfidf_matrix, index_to_name, name_to_index = build_tfidf_matrix(perfumes_df)
    print("name to index",name_to_index)

    # Suppose the user selected these
    user_selected = ["Le Male Pride 2023", "Classique Pride 2024"]

    # C) Get recommendations
    # We do top_k_each=200 => each item fetch top 200, final = top 5
    recs = get_content_based_recommendations(
        selected_perfumes=user_selected,
        tfidf_matrix=tfidf_matrix,
        name_to_index=name_to_index,
        index_to_name=index_to_name,
        top_k_each=200,
        final_n=5
    )
    print("User selected:", user_selected)
    print("Recommendations:", recs)
```

```python
#############################################
# rowwise_collab_recommender.py
#############################################

import logging
import pandas as pd
import numpy as np

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(name)s - %(message)s'
)
logger = logging.getLogger(__name__)

def build_item_user_matrix(
    users_csv,
    perfumes_csv,
    user_col="UserID",
    delimiter_users=",",
    delimiter_perfumes=";",
    encoding_users="latin-1",      # Change to match your users.csv
    encoding_perfumes="latin-1"   # Change to match your perfumes_updated.csv
):
    """
    Reads 'users.csv' and 'perfumes_updated.csv' with specified encodings.
    Builds item_user matrix in row-by-row approach.
    """
    logger.info("Loading user data from %s", users_csv)
    df_users = pd.read_csv(users_csv, delimiter=delimiter_users, encoding=encoding_users)

    if user_col not in df_users.columns:
        logger.warning("No '%s' column found in users CSV. We'll use row index as user ID.", user_col)
        df_users[user_col] = df_users.index

    user_id_to_perfumes = {}
    for idx, row in df_users.iterrows():
        user_id = row[user_col]
        raw_list = str(row.get("Perfumes", "")).split(",")
        chosen = {p.strip().lower() for p in raw_list if p.strip()}
        user_id_to_perfumes[user_id] = chosen

    logger.info("Loading perfume data from %s", perfumes_csv)
    df_perfumes = pd.read_csv(perfumes_csv, delimiter=delimiter_perfumes, encoding=encoding_perfumes)

    all_perfume_names = df_perfumes["Perfume"].unique().tolist()
    perfume_lower_to_original = {p.lower(): p for p in all_perfume_names}

    logger.info("Building user-item matrix. #users=%d, #perfumes=%d", len(user_id_to_perfumes), len(all_perfume_names))

    item_names = sorted(all_perfume_names)
    user_names = sorted(user_id_to_perfumes.keys())

    # Create user-item (#users x #perfumes) filled with 0
    user_item_df = pd.DataFrame(
        data=0,
        index=user_names,
        columns=item_names,
        dtype=np.float32
    )

    # Fill 1 if user selected that perfume
    for user_id, chosen_set in user_id_to_perfumes.items():
        for c in chosen_set:
            if c in perfume_lower_to_original:
                real_name = perfume_lower_to_original[c]
                if real_name in user_item_df.columns:
                    user_item_df.at[user_id, real_name] = 1.0

    # Transpose => #perfumes x #users
    item_user_df = user_item_df.T
    logger.info("item_user matrix shape = %s", item_user_df.shape)

    item_to_index = {name: i for i, name in enumerate(item_user_df.index)}
    return item_user_df, item_user_df.index.tolist(), item_user_df.columns.tolist(), item_to_index


def get_topk_for_item(item_name, item_user_df, item_to_index, top_k=50):
    item_name_lower = item_name.lower()
    matched_name = None
    for real_name, idx in item_to_index.items():
        if real_name.lower() == item_name_lower:
            matched_name = real_name
            break
    if matched_name is None:
        logging.warning("Item '%s' not found. Returning [].", item_name)
        return []

    row_i = item_user_df.loc[matched_name].values.reshape(1, -1)
    matrix_values = item_user_df.values
    sim_vec = row_i @ matrix_values.T
    sim_vec = sim_vec.ravel()

    self_idx = item_to_index[matched_name]
    sim_vec[self_idx] = -999.0

    n_items = len(sim_vec)
    if top_k >= n_items:
        top_indices = np.argsort(sim_vec)[::-1][:top_k]
    else:
        partition_idx = np.argpartition(sim_vec, -top_k)[-top_k:]
        top_indices = partition_idx[np.argsort(sim_vec[partition_idx])[::-1]]

    all_names = item_user_df.index
    results = [(all_names[idx], float(sim_vec[idx])) for idx in top_indices]
    return results


def get_collab_recommendations(selected_items, item_user_df, item_to_index, top_k_each=50, final_n=5):
    logger.info("Starting item-based collab aggregator for %s", selected_items)
    agg_scores = {}
    selected_lower = {si.lower() for si in selected_items}

    for itm in selected_items:
        topk_neighbors = get_topk_for_item(itm, item_user_df, item_to_index, top_k=top_k_each)
        for (neighbor_name, score) in topk_neighbors:
            if neighbor_name.lower() in selected_lower:
                continue
            agg_scores[neighbor_name] = agg_scores.get(neighbor_name, 0.0) + score

    if not agg_scores:
        logger.warning("No aggregated scores found. Possibly mismatch in naming or no data.")
        return []

    sorted_items = sorted(agg_scores.items(), key=lambda x: x[1], reverse=True)
    recs = [name for (name, score) in sorted_items[:final_n]]
    logger.info("Final recommendations => %s", recs)
    return recs


if __name__ == "__main__":
    item_user_df, item_names, user_names, item_to_index = build_item_user_matrix(
        users_csv="users.csv",
        perfumes_csv="perfumes_updated.csv",
        user_col="UserID",
        delimiter_users=",",
        delimiter_perfumes=";",
        encoding_users="latin-1",      # Adjust if needed
        encoding_perfumes="latin-1"   # Adjust if needed
    )

    selected_perfumes = ["Le Male Pride 2023", "Classique Pride 2024"]
    recs = get_collab_recommendations(
        selected_items=selected_perfumes,
        item_user_df=item_user_df,
        item_to_index=item_to_index,
        top_k_each=200,
        final_n=5
    )
    print("User selected:", selected_perfumes)
    print("Recommendations:", recs)
```
